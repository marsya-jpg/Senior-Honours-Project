```python
file = open("Eagle_total_stellar_mass.txt")
import matplotlib.pyplot as plt
# [0] redshift z, [1] total stellar mass in 10^10 M_sun/h

rho_c = 22.7754*10**10 #h^2 Msun Mpc^-3
L = 100 #Mpc
omega_z = []
z = []

for i in file:
    val = file.readlines() #Lists each star values in a list
    for line in val: #Splitting redshift and stellar mass
        data = line.split(" ")
        redshift = float(data[0])
        mass = float(data[1])   #10^10 h^-1 Msun
        rho = mass*100**3 # multiply Box size for volume
        omega_z.append(rho/rho_c)
        z.append(redshift)
        
with open('EAGLE_omega_output.txt', 'a') as myfile:
    for z_data, omega_data in zip(z, omega_z):
        myfile.write("%s %s\n"%(z_data, omega_data))
    
        
plt.scatter(z, omega_z, 2)
plt.xlabel('redshift, z')
plt.ylabel('omega_*')
plt.title("EAGLE: omega* against redshift")
plt.show()
```

```python
file = open("Illustris_total_stellar_mass.txt")
import matplotlib.pyplot as plt
# [0] redshift z, [1] total stellar mass in 10^10 M_sun/h

rho_c = 22.7754*10**10 
L = 75 #Mpc
omega_z = []
z = []

for i in file:
    val = file.readlines() #Lists each star values in a list
    for line in val: #Splitting redshift and stellar mass
        data = line.split(" ")
        redshift = float(data[1])
        mass = float(data[2])
        rho = mass*110.7**3
        omega_z.append(rho/rho_c)
        z.append(redshift)

with open('Illustris_omega_output.txt', 'a') as myfile:
    for z_data, omega_data in zip(z, omega_z):
        myfile.write("%s %s\n"%(z_data, omega_data))

plt.scatter(z, omega_z, 2)
plt.xlabel('redshift, z')
plt.ylabel('omega_*')
plt.title("Illustris: omega* against redshift")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid.axislines import SubplotZero


#Opening lists
z1 = []
omega1 = []
z2 = []
omega2 = []
z3 = []
omega3 = []

#EAGLE
with open("EAGLE_omega_output.txt") as file1:
    val = file1.readlines() #Lists each star values in a list
    for line in val: #Splitting redshift and stellar mass
        data = line.split(" ")
        data_z = float(data[0])
        data_omega = float(data[1])
        z1.append(data_z)
        omega1.append(data_omega)

#Illustris
with open("Illustris_omega_output.txt") as file2:
    val = file2.readlines() #Lists each star values in a list
    for line in val: #Splitting redshift and stellar mass
        data = line.split(" ")
        data_z = float(data[0])
        data_omega = float(data[1])
        z2.append(data_z)
        omega2.append(data_omega)
        
#SIMBA
with open("SIMBA_omega_output.txt") as file3:
    val = file3.readlines() #Lists each star values in a list
    for line in val: #Splitting redshift and stellar mass
        data = line.split(" ")
        data_z = float(data[0])
        data_omega = float(data[1])
        z3.append(data_z)
        omega3.append(data_omega)

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(z1, omega1,2, c="blue")
ax.scatter(z2, omega2,2, c="red")
ax.scatter(z3, omega3,3, c="orange")
#ax.set_xlim([0,14])
#ax.set_ylim([-0.01,0.1])
plt.xlabel('redshift, z')
plt.ylabel('omega_*')
plt.title(" omega* against redshift")
plt.legend(['EAGLE', 'Illustris','SIMBA'])
plt.yscale("log")
plt.show()    



#want to know max omega* at z~0
print(max(omega1), "EAGLE y-intercept")
print(max(omega2), "Illustris y-intercept")
print(max(omega3), "SIMBA y-intercept")
```
